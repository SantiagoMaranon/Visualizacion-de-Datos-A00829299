```python
import pandas as pd
import re
import numpy as np
import seaborn as sns

df = pd.read_csv("datasets/pokemon.csv")
```

```python
df
```

```python
T = pd.pivot_table (df, index = "Name", values = ["Defense","Attack"])
```

```python
print(T)
```

```python
sns.relplot(data = T, x = "Defense", y = "Attack")
```

```python
df.rename(columns={'HP':'Salud', 'Attack':'Ataque','Defense':'Defensa'}, inplace=True)
df
```

```python
df_Tipo1 = df.groupby('Type 1', 
                           axis=0).sum()

df_Tipo1.head()
```

```python
df_Tipo1['Total'].plot(kind='pie',
                            figsize=(5, 6),
                            startangle=90,     
                            shadow=True,           
                            )

```
